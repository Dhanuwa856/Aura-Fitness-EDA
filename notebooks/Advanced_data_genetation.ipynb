# üìù Unit 13.9: Advanced Correlated Dataset (‡∑É‡∑í‡∂∫‡∂Ω‡∑î ‡∑É‡∂∏‡∑ä‡∂∂‡∂±‡∑ä‡∂∞‡∂≠‡∑è ‡∑É‡∑Ñ‡∑í‡∂≠‡∑Ä)

```python
import pandas as pd
import numpy as np
```

```python
# 1. ‡∂∏‡∑ñ‡∂Ω‡∑í‡∂ö ‡∑É‡∑ê‡∂ö‡∑É‡∑î‡∂∏‡∑ä
num_rows = 5000
np.random.seed(42)

# 2. ‡∑É‡∑ä‡∑Ä‡∑è‡∂∫‡∂≠‡∑ä‡∂≠ ‡∂Ø‡∂≠‡∑ä‡∂≠ (Independent Variables)
age = np.random.randint(18, 65, size=num_rows)
gender = np.random.choice(['male', 'female'], size=num_rows)
workout_min = np.random.randint(10, 150, size=num_rows).astype(float)
```

```python
# 3. ‡∑É‡∂∏‡∑ä‡∂∂‡∂±‡∑ä‡∂∞‡∂≠‡∑è‡∑Ä‡∂∫‡∂±‡∑ä ‡∂ú‡∑ú‡∂©‡∂±‡∑ê‡∂ú‡∑ì‡∂∏ (Building the Web of Correlations)

# Daily_Steps: Workout_Min ‡∂∏‡∂≠ ‡∂ª‡∂≥‡∑è ‡∂¥‡∑Ä‡∂≠‡∑ì
daily_steps = (workout_min * 85) + np.random.normal(1000, 500,size=num_rows)
daily_steps
```

```python
# Calories_Burned: Steps ‡∑É‡∑Ñ Workout ‡∂Ø‡∑ô‡∂ö‡∂∏ ‡∂∏‡∂≠ ‡∂ª‡∂≥‡∑è ‡∂¥‡∑Ä‡∂≠‡∑ì

calories_burned = (daily_steps * 0.04 ) + (workout_min * 4) + np.random.normal(50,20,size=num_rows)
calories_burned
```

```python
# Water_Intake: Steps ‡∑Ä‡∑ê‡∂©‡∑í ‡∑Ä‡∂± ‡∑Ä‡∑í‡∂ß ‡∑É‡∑í‡∂ª‡∑î‡∂ª‡∂ß ‡∑Ä‡∑ê‡∂©‡∑í‡∂¥‡∑î‡∂ª ‡∂¢‡∂Ω‡∂∫ ‡∂Ö‡∑Ä‡∑Å‡∑ä‚Äç‡∂∫ ‡∑Ä‡∑ö
water_intake = (daily_steps * 0.0002) + (workout_min * 0.01) + np.random.normal(1,0.5,size=num_rows)
water_intake
```

```python
# Protein_Intake: Workout ‡∑Ä‡∑ê‡∂©‡∑í ‡∑Ä‡∂± ‡∑Ä‡∑í‡∂ß ‡∂¥‡∂ß‡∂ö ‡∂Ö‡∂Ω‡∑î‡∂≠‡∑ä‡∑Ä‡∑ê‡∂©‡∑í‡∂∫‡∑è‡∑Ä‡∂ß ‡∂¥‡∑ä‚Äç‡∂ª‡∑ù‡∂ß‡∑ì‡∂±‡∑ä ‡∂Ö‡∑Ä‡∑Å‡∑ä‚Äç‡∂∫ ‡∑Ä‡∑ö
protein_intake = (workout_min * 0.5) + (age * 0.2) + np.random.normal(20,5 ,size=num_rows)
protein_intake
```

```python
# Sleep_Hrs: ‡∂∏‡∂∞‡∑ä‚Äç‡∂∫‡∑É‡∑ä‡∂Æ ‡∑Ä‡∑ä‚Äç‡∂∫‡∑è‡∂∫‡∑è‡∂∏ ‡∂±‡∑í‡∂±‡∑ä‡∂Ø ‡∑Ä‡∑ê‡∂©‡∑í ‡∂ö‡∂ª‡∂± ‡∂Ö‡∂≠‡∂ª ‡∂Ö‡∂∞‡∑í‡∂ö ‡∑Ä‡∑ä‚Äç‡∂∫‡∑è‡∂∫‡∑è‡∂∏ ‡∂±‡∑í‡∂±‡∑ä‡∂Ø ‡∂Ö‡∂©‡∑î ‡∂ö‡∑Ö ‡∑Ñ‡∑ê‡∂ö (‡∂∞‡∂± ‡∑É‡∂∏‡∑ä‡∂∂‡∂±‡∑ä‡∂∞‡∂∫‡∂ö‡∑ä ‡∂ú‡∂∏‡∑î)

sleep_hrs = (daily_steps * 0.0001) + np.random.normal(6.5,0.8,size=num_rows)
sleep_hrs
```

```python
# 4. ‡∂Ø‡∂≠‡∑ä‡∂≠ ‡∂¥‡∑í‡∂ª‡∑í‡∑É‡∑í‡∂Ø‡∑î ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏ (‡∑É‡∑ì‡∂∏‡∑è‡∑Ä‡∂±‡∑ä ‡∂¥‡∑Ä‡∂≠‡∑ä‡∑Ä‡∑è ‡∂ú‡∑ê‡∂±‡∑ì‡∂∏)
daily_steps = np.clip(daily_steps, 500, 20000).round(0)
calories_burned = np.clip(calories_burned, 100, 1500).round(0)
water_intake = np.clip(water_intake, 1.0, 6.0).round(1)
protein_intake = np.clip(protein_intake, 20, 150).round(0)
sleep_hrs = np.clip(sleep_hrs, 4, 10).round(1)
```

```python
data = {
    'User_ID': [f'U{i:04d}' for i in range(1, num_rows + 1)],
    'Age': age,
    'Gender': gender,
    'Daily_Steps': daily_steps,
    'Workout_Min': workout_min,
    'Calories_Burned': calories_burned,
    'Sleep_Hrs': sleep_hrs,
    'Protein_Intake': protein_intake,
    'Water_Intake': water_intake
}
```

```python
df_aura_final = pd.DataFrame(data)
```

```python
# 3. ‡∑Ñ‡∑í‡∂≠‡∑è‡∂∏‡∂≠‡∑è‡∂∏ NaN values ‡∂á‡∂≠‡∑î‡∑Ö‡∂≠‡∑ä ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏ (5% missing values)
for col in ['Daily_Steps', 'Workout_Min', 'Calories_Burned', 'Sleep_Hrs','Protein_Intake', 'Water_Intake']:
    mask = np.random.random(num_rows) < 0.05
    df_aura_final.loc[mask, col] = np.nan

df_aura_final.tail()
```

```python
# 6. Save ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏ (Root data folder ‡∂ë‡∂ö‡∂ß)
df_aura_final.to_csv('../data/aura_fitness_final.csv', index=False)
print("Successful! The 'aura_fitness_final.csv' file with all connections was created. üöÄ")
```
