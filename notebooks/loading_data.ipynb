```python
# ‡∂ë‡∂ö‡∑ä ‡∂ë‡∂ö‡∑ä column ‡∂ë‡∂ö‡∑ô‡∑Ñ‡∑í ‡∂á‡∂≠‡∑í ‡∂∏‡∑î‡∑Ö‡∑î ‡∑Ñ‡∑í‡∑É‡∑ä ‡∂≠‡∑ê‡∂±‡∑ä ‡∂ú‡∂´‡∂± ‡∂∂‡∑ê‡∂Ω‡∑ì‡∂∏
print("Amount of empty space before cleaning:\n")
print(df.isnull().sum())
```

```python
# Numerical columns ‡∂Ω‡∑ê‡∂∫‡∑í‡∑É‡∑ä‡∂≠‡∑î‡∑Ä
cols_to_fix = ['Daily_Steps', 'Workout_Min', 'Sleep_Hrs', 'Protein_Intake']

# Loop ‡∂ë‡∂ö‡∂ö‡∑ä ‡∂∏‡∂ú‡∑í‡∂±‡∑ä ‡∂ë‡∂ö‡∑Ä‡∂ª ‡∂¥‡∑í‡∂ª‡∑í‡∑É‡∑í‡∂Ø‡∑î ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏

for col in cols_to_fix:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
```

```python
# ‡∂¥‡∑í‡∂ª‡∑í‡∑É‡∑í‡∂Ø‡∑î ‡∂ö‡∑Ö ‡∂¥‡∑É‡∑î ‡∂±‡∑ê‡∑Ä‡∂≠ ‡∂¥‡∂ª‡∑ì‡∂ö‡∑ä‡∑Ç‡∑è ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏
print("\nAmount of empty space after cleaning:\n")
print(df.isnull().sum())
```
