```python
import numpy as np
import pandas as pd
import os
```

```python
# 1. 'data' folder ‡∂ë‡∂ö ‡∂±‡∑ê‡∂≠‡∑ä‡∂±‡∂∏‡∑ä ‡∂í‡∂ö ‡∑Ñ‡∂Ø‡∂±‡∑ä‡∂±
if not os.path.exists('../data'):
    os.mkdir('../data')

# 2. ‡∂¥‡∑ö‡∑Ö‡∑í ‡∂ú‡∂´‡∂± 5000 ‡∂Ω‡∑ô‡∑É ‡∑É‡∂ö‡∑É‡∂∏‡∑î
num_rows = 5000
np.random.seed(42) # ‡∑Ñ‡∑ê‡∂∏‡∂≠‡∑í‡∑É‡∑ä‡∑É‡∑ô‡∂∏ ‡∂ë‡∂ö‡∂∏ ‡∂Ø‡∂≠‡∑ä‡∂≠ ‡∂Ω‡∑ê‡∂∂‡∑ô‡∂±‡∑ä‡∂± (Reproducibility)

data = {
    'User_ID':[f'U{i:04d}' for i in range(1,num_rows+ 1)],
    'Age': np.random.randint(18,65,size=num_rows),
    'Gender': np.random.choice(['male','female'],size=num_rows),
    'Daily_Steps':np.random.randint(500,1800,size=num_rows),
    'Workout_Min':np.random.randint(0,150,size=num_rows),
    'Sleep_Hrs':np.random.randint(4,10,size=num_rows).round(1),
    'Protein_Intake':np.random.randint(10,120,size=num_rows),
    'Water_Intake':np.random.uniform(0.5,5.0,size=num_rows).round(1)
}

df_aura = pd.DataFrame(data)

# 3. CSV ‡∂ë‡∂ö Save ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏
file_path = '../data/aura_fitness_data.csv'

df_aura.to_csv(file_path, index=False)
print(f"Success! File '{file_path}' with {num_rows} rows created. ‚úÖ")
```
